# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

customers_scale = StandardScaler().fit_transform(customers)
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from scipy.cluster.hierarchy import dendrogram, linkage

plt.figure(figsize=(15, 8))

Z=linkage(customers_scale, 'ward')

dendrogram(Z, truncate_mode='lastp', p=12,
           leaf_rotation=90., leaf_font_size=8.,
           show_contracted=True, show_leaf_counts=False)

plt.title('Dendrograma')
plt.xlabel('indice de muestra')
plt.ylabel('distancia')

plt.show();
```

```python
def fancy_dendro(*args, **kwargs):
    
    max_d=kwargs.pop('max_d', None)
    
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold']=max_d
        
    annotate_above=kwargs.pop('annotate_above', None)
    
    ddata=dendrogram(*args, **kwargs)
    
    if not kwargs.get('no_plot', False):
        plt.title('Dendrograma')
        plt.xlabel('Indice de muestra o (tamaño del cluster)')
        plt.ylabel('Distancia')
        
        for i,d,c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x=0.5*sum(i[1:3])
            y=d[1]
            
            if y>annotate_above:
                plt.plot(x,y,'o', c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),textcoords='offset points', va='top', ha='center')
                
        if max_d:
            plt.axhline(y=max_d, c='k')
    
    return ddata
    
    
```

```python
fancy_dendro(Z, truncate_mode='lastp', p=12,
                 leaf_rotation=90., leaf_font_size=8.,
                 show_contracted=True, annotate_above=10, 
             
             max_d=200000)

plt.show();
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3) #El dendongrama nos ha dado a entender que los datos se pueden clusterizar en 3

clusters = kmeans.fit(customers_scale)
```

```python
len(clusters.cluster_centers_)
```

```python
preds = clusters.fit_predict(customers_scale)

preds[0:10]
```

Count the values in `labels`.

```python
len(clusters.labels_)
```

```python
customers['labels_Kmeans'] = preds
```

```python
customers['labels_Kmeans'].value_counts()
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
kmeans=KMeans(n_clusters=2)

clusters3 = kmeans.fit(customers_scale)
```

```python
customers['labels_Kmeans_2'] = clusters3.labels_
```

```python
fig, axs = plt.subplots(1, 2, sharey=True)

axs[0].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_Kmeans_2)     # color
axs[0].set_title('2 clusters')

axs[1].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_Kmeans) # color
axs[1].set_title('3 clusters')

plt.show()
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
dbscan=DBSCAN(eps=0.5, min_samples = 10)

clusters4 = dbscan.fit(customers_scale)

customers['labels_DBSCAN2'] = clusters4.labels_
```

```python
dbscan=DBSCAN(eps=0.5, min_samples = 30)

clusters5 = dbscan.fit(customers_scale)

customers['labels_DBSCAN3'] = clusters5.labels_
```

```python
dbscan=DBSCAN(eps=0.1)

clusters6 = dbscan.fit(customers_scale)

customers['labels_DBSCAN4'] = clusters6.labels_
```

```python
dbscan=DBSCAN(eps=0.9)

clusters7 = dbscan.fit(customers_scale)

customers['labels_DBSCAN5'] = clusters7.labels_
```

```python
fig, axs = plt.subplots(2, 3, sharey=True, sharex=True)

axs[0][0].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_DBSCAN)     # color
axs[0][0].set_title('eps=0.5')

axs[0][1].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_DBSCAN2) # color
axs[0][1].set_title('eps=0.5, min_samples = 10')

axs[0][2].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_DBSCAN3)     # color
axs[0][2].set_title('eps=0.5, min_samples = 30')

axs[1][0].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_DBSCAN) # color
axs[1][0].set_title('eps=0.5')

axs[1][1].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_DBSCAN4)     # color
axs[1][1].set_title('eps=0.1')

axs[1][2].scatter(customers.Frozen,   # x
            customers.Delicassen,   # y
            c=customers.labels_DBSCAN5) # color
axs[1][2].set_title('eps=0.9')

plt.show()
```

**Your comment here**




eps: este parámetro define la distancia máxima entre dos puntos para que sean considerados vecinos y, por lo tanto, sean parte del mismo grupo. Los puntos que están a una distancia menor o igual a eps se consideran vecinos. Si el valor de eps es demasiado pequeño, puede haber muchos grupos pequeños; si es demasiado grande, todos los puntos pueden formar un solo grupo. Por lo tanto, el valor de eps debe ajustarse cuidadosamente en función de la densidad de los datos.

min_samples: este parámetro establece la cantidad mínima de puntos que deben estar dentro de una distancia de eps para que se considere que un punto es un punto central o "core" y, por lo tanto, forme parte de un grupo. En otras palabras, un punto debe tener al menos min_samples vecinos cercanos para ser considerado parte de un grupo. Si el valor de min_samples es demasiado pequeño, se pueden formar grupos muy pequeños y ruidosos; si es demasiado grande, pueden haber muchos puntos que no se asignen a ningún grupo. Por lo tanto, el valor de min_samples también debe ajustarse cuidadosamente en función de la densidad de los datos.
